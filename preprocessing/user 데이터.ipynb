```python
dev_users = []
path1 = '/Volumes/Seoin Baek/브런치 사용자를 위한 글 추천 대회/res/predict/dev.users.txt'
with open(path1,'r') as dev:
    for line in dev:
        line = line.rstrip()
        dev_users.append(line)
```

```python
test_users = []
path2 = '/Volumes/Seoin Baek/브런치 사용자를 위한 글 추천 대회/res/predict/test.users.txt'
with open(path2,'r') as test:
    for line in test:
        line = line.rstrip()
        test_users.append(line)
print(len(test_users))
```

```python
import json
import ujson
from pprint import pprint

USER = dict()
check = dict()
if __name__ == "__main__":
    file_path = '/Volumes/Seoin Baek/브런치 사용자를 위한 글 추천 대회/users.json.txt'
    with open(file_path,'r') as file_name:
        for users in file_name:
            users = ujson.loads(users)
            if users["id"] in test_users:
                for i in users["following_list"]:
                    try:
                        USER[users["id"]][i] = 1
                    except:
                        USER[users["id"]] = dict()
                        USER[users["id"]][i] = 1 
                    try:
                        check[i] +=1
                    except:
                        check[i] = 1
```

```python
for i in test_users:
    i = i.rstrip()
    if i not in USER:
        USER[i] = dict()
```
